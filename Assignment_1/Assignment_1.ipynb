I can not remove outliers/anomalies freely as they might have a strong impact on output.

```python
# Check the distribution of Financial Distress first
# if value>-0.5 healthy, aka a 0, if value <=-0.5 unhealthy, aka 1
np.bincount(data['Financial Distress'] > -0.5)
```

```python
def outliers_iqr(ys):
    quartile_1, quartile_3 = np.percentile(ys, [15, 85]) 
    iqr = quartile_3 - quartile_1
    lower_bound = quartile_1 - (iqr * 1.5)
    upper_bound = quartile_3 + (iqr * 1.5)
    return np.where((ys > upper_bound) | (ys < lower_bound))
```

```python
outlier_list = []
for column_index in list(range(1,80)) + list(range(81,83)):
    outlier_list = outlier_list+ outliers_iqr(data['x' + str(column_index)])[0].tolist()
test_d = dict([i, outlier_list.count(i)] for i in outlier_list)
print('Number of rows has outliers is: ' + str(len(test_d))) 
```

```python
a = 0
b = 0
for x in list(test_d.keys()):
    if data.iloc[x,2]>-0.5:
        a = a + 1
    else: b = b + 1
print('Number of healthy records with outliers is: ' + str(a)) 
print('Number of healthy records with outliers is: ' + str(b)) 
```

### Data preprocessing

```python
from patsy import dmatrices
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn import preprocessing
import statsmodels.api as sm
```

1. one hot coding. Convert categorical variables to numeric
2. normalization
3. reduce demsion by vif and linear analysis

```python
#Remove first two columns as they are irrelavent, also rename first column
data_preprocess = data.drop(data.columns[[0,1]], axis=1, inplace=False)
data_preprocess.rename(columns={'Financial Distress':'Financial_Distress'}, inplace=True)
data_preprocess.head(3)
```

```python
#Check how many categories for x80
x80_cats = data_preprocess['x80'].value_counts()
len(x80_cats)
```

```python
#Use one hot coding to encode the categorical variable
data_preprocess = pd.get_dummies(data_preprocess, columns=["x80"])
data_preprocess.head()
```

```python
#Normalization
x = data_preprocess.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
data_preprocess = pd.DataFrame(x_scaled, columns=data_preprocess.columns)
data_preprocess.head(3)
```

```python
#VIF
#gather features
features = "+".join(data_preprocess.drop(['Financial_Distress'], axis=1))
# get y and X dataframes based on this regression:
y, X = dmatrices('Financial_Distress ~' + features, data_preprocess, return_type='dataframe')

# For each X, calculate VIF and save in dataframe
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["features"] = X.columns
```

```python
vif_sort = vif.sort_values('VIF Factor', ascending=0)
vif_sort[vif_sort['VIF Factor']==np.inf].sort_values('features')
```

```python
#For the ones equal to inf，only keep column x80_1
inf_cols = vif_sort[vif_sort['VIF Factor']==np.inf]['features'].sort_values().values.tolist()[2:]
data_preprocess.drop(inf_cols, inplace=True, axis=1)
data_preprocess.head()
```

```python
data_preprocess.shape
```

```python
vif_sort[vif_sort['VIF Factor']!=np.inf]
```

```python
#Get ride of top 15 columns as they have extremely large vif values
dul_cols = vif_sort[vif_sort['VIF Factor']!=np.inf]['features'][:15]
data_preprocess.drop(dul_cols, inplace=True, axis=1)
print(data_preprocess.shape)
data_preprocess.head()
```

```python
#70 columns left, run a multiple linear analysis 
X = data_preprocess.drop(['Financial_Distress'], axis=1)
y = data_preprocess.Financial_Distress

X2 = sm.add_constant(X)
est = sm.OLS(y, X2)
est2 = est.fit()
print(est2.summary())
```

```python
#Keep columns with p value <= 0.05
new_cols = ['Financial_Distress'] + est2.pvalues[est2.pvalues <= 0.05].index.tolist()
```

```python
print(data_preprocess[new_cols].shape)
data_preprocess[new_cols].head(3)
```

### Part B:  Analyze Some Data

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score, make_scorer
from sklearn import svm
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression
from sklearn.neural_network import MLPRegressor
import random
```

```python
random.seed(0)
y = data_preprocess['Financial_Distress'] #Define dependent variable as y
df = data_preprocess.drop(data_preprocess.columns[[0]], axis=1, inplace=False) #Define independent variables x1-x83 as df
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.3)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
def MSE(y_true,y_pred):
    mse = mean_squared_error(y_true, y_pred)
    return mse
    
def R2(y_true,y_pred):  
    #Best possible score is 1, if it's negative, then the model perform worse than a horizontal line.
    r2 = r2_score(y_true, y_pred) 
    return r2
def two_score(y_true,y_pred):    
    MSE(y_true,y_pred) #set score here and not below if using MSE in GridCV
    score = R2(y_true,y_pred)
    return score

def two_scorer():
    return make_scorer(two_score, greater_is_better=True)
```

#### Multiple linear regression

```python
mlinear_model = linear_model.LinearRegression()
lm = mlinear_model.fit(X_train, y_train)
mlinear_model.predict(X_test)
```

```python
# The coefficients
print('Coefficients: \n', mlinear_model.coef_)
print('Mean squared error: %.2f' % MSE(y_test, mlinear_model.predict(X_test))) #...bad
print('Variance score: %.2f' % R2(y_test, mlinear_model.predict(X_test))) #...this is horrible
```

#### SVR

```python
svr_model = svm.SVR(gamma=0.001, C=100, epsilon = 0.001)
svr_model.fit(X_train, y_train) 
```

```python
print('Mean squared error: %.2f' % MSE(y_test, svr_model.predict(X_test))) #good!
print('Variance score: %.2f' % R2(y_test, svr_model.predict(X_test))) #fine
```

```python
def svr_param_selection(X, y, nfolds):
    tuned_parameters = {'kernel': ['rbf'], 'gamma': [0.001, 0.01, 0.1, 1],  
                     'C': [1, 5, 10, 100, 1000], 'epsilon' : [0.001, 0.01, 0.1]}
    grid_search = GridSearchCV(svm.SVR(), tuned_parameters, cv=nfolds, scoring=two_scorer())
    grid_search.fit(X, y)
    grid_search.best_params_
    return grid_search.best_params_
```

```python
svr_param_selection(X_train, y_train, 5)
```

```python
svr_model_2 = svm.SVR(kernel='rbf', C=1, gamma=0.1, epsilon=0.001)
svr_model_2.fit(X_train, y_train) 
```

```python
print('Mean squared error: %.2f' % MSE(y_test, svr_model_2.predict(X_test)))
print('Variance score: %.2f' % R2(y_test, svr_model_2.predict(X_test))) #better
```

```python
def rf_param_selection(X, y, nfolds):
    tuned_parameters = {'max_depth': range(2,6), 
                        'n_estimators':[1,5, 10,100],
                        'max_features' : ['auto', 'log2', 'sqrt'],
                        'random_state' : [0,1]} 
    grid_search = GridSearchCV(RandomForestRegressor(), tuned_parameters, cv=nfolds, scoring=two_scorer())
    grid_search.fit(X, y)
    grid_search.best_params_
    return grid_search.best_params_
```
